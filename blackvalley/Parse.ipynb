```python
import json
import re

with open('data.json') as data_file:
    data = json.load(data_file)
```

```python
members = data['response']['members']
unclean = [re.split(r'[\|\-\/]',m['nickname'].lower().encode('ascii','ignore')) for m in members]

# Create a dict to find most common occurences
result = {}
for instance in unclean:
    for i, string in enumerate(instance):
        
        # Skip what is in most instance a name
        if i == 0:
            continue
        
        # Resulting key is likely university or company
        key = string.strip()
        if string in result:
            result[key] += 1
        else:
            result[key] = 1
```
